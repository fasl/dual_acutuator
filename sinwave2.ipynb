```python
import wave
import numpy as np
import matplotlib.pyplot as plt

# WAV file parameters
num_channels = 4
sample_width = 2  # 2 bytes for 16-bit audio
sample_rate = 44100
duration = 5  # seconds

# Generate dummy 4-channel audio data (e.g., sine waves for each channel)
num_samples = int(sample_rate * duration)
t = np.linspace(0, duration, num_samples, endpoint=False)

# Create a 4-channel array
# Each row represents a frame, each column represents a channel
audio_data = np.zeros((num_samples, num_channels), dtype=np.int16)

# Populate each channel with a different frequency sine wave
frequency_ch1 = 440  #
frequency_ch2 = 440  #
frequency_ch3 = 440  #
frequency_ch4 = 440  #

phase = 0.5  # isousa
amplitude = 10000  # Max amplitude for int16

# define func
def sinfunc(amplitude,t,phase):
   return (amplitude * np.sin(2 * np.pi * frequency_ch2 * t + phase * np.pi)).astype(np.int16)

# calculate wave
audio_data[:, 0] = sinfunc(amplitude,t,0)
audio_data[:, 1] = sinfunc(amplitude,t,phase)
audio_data[:, 2] = sinfunc(amplitude,t,0)
audio_data[:, 3] = sinfunc(amplitude,t,phase)
# audio_data[:, 1] = sinfunc(amplitude * np.sin(2 * np.pi * frequency_ch2 * t + phase * np.pi)).astype(np.int16)
```

```python
t = np.linspace(0, 1, 50)
y1 =  sinfunc(amplitude,t,0)
y2 =  sinfunc(amplitude,t,phase)

plt.plot(t,y1)
plt.plot(t,y2)

# Add labels and title
plt.xlabel("time")
plt.ylabel("Y-axis")
plt.title("Wave Plot")

# Display the plot
plt.show()
```

```python


# Open and write the WAV file
output_filename = "sinwave_0_5pie_440.wav"
with wave.open(output_filename, 'wb') as wavfile:
    wavfile.setnchannels(num_channels)
    wavfile.setsampwidth(sample_width)
    wavfile.setframerate(sample_rate)
    wavfile.writeframes(audio_data.tobytes())

print(f"4-channel WAV file '{output_filename}' created successfully.")
```
